IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

PROCESSING DATA

```python
data = pd.read_csv("pokemon.csv")
x = data.Attack
y = data.HP
plt.xlabel("Attack")
plt.ylabel("HP")
plt.scatter(x,y)
plt.show()
```

BUILDING MODEL

```python
#y = mx+c
m=0
c=0

#learning rate
L = 0.0001

#number of operations to perform on gradient descent
epochs = 5000

n = float(len(x))
# Performing Gradient Descent 
for i in range(epochs): 
    y_pred = m*x + c  # The current predicted value of Y
    D_m = (-2/n) * sum(x * (y - y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(y - y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c

print (m, c)
```

MAKING PREDICTIONS

```python
y_pred = m*x+c
plt.scatter(x,y)
# regression line
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color='red')  
plt.show()
```
